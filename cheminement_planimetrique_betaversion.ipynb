```python
#calcul du cheminement 
import numpy as np 
import pandas as pd 
from utils import *
```

```python
## need to import the files 
file = pd.read_excel(r'test23.xlsx' ,nrows = 30)
file
```

```python
angles = np.zeros(int(len(file.rcg) / 3 ))
rcgg = np.zeros(len(file.rcg))
rcdd = np.zeros(len(file.rcg))
cd = file.cd
cg = file.cg
moyenne = np.zeros(len(file.rcg)) 
n = len(file.rcg) 
for i in range(1 , n  , 3 ) : 
    p = file.cg[i+1] - file.cg[i-1]
    if p> 399  :
        rcgg[i] = roundd(file.cg[i] - (file.cg[i+1]+400+file.cg[i-1])/2)
    else: 
        rcgg[i] =  roundd(file.cg[i] - (file.cg[i+1] +file.cg[i-1])/2)
for i in range(1 , n  , 3 ) : 
    p = file.cd[i+1] - file.cd[i-1]
    if p> 399  :
        rcdd[i] =  roundd(file.cd[i] - (file.cd[i+1]+400+file.cd[i-1])/2)
    else: 
        rcdd[i] =  roundd(file.cd[i] - (file.cd[i+1] +file.cd[i-1])/2)
    moyenne[i] = ((rcdd[i] + rcgg[i]) / 2 )  
    angles[i//3] = moyenne[i]
file.rcg = rcgg 
file.rcd = rcdd 
file.moyenne = moyenne
file
```

```python
n = 1  # nombre de serie par mesure rops
angles_finales = np.zeros(int(len(angles) / n )) # nombre de serie par mesure  
for i in range(0 , len(angles) , n )  : 
    somme = 0 
    for j in range(i , i + n ) : 
        somme  = somme + angles[j]
    
    angles_finales[i//n] = somme / n 
print(angles_finales)
```

```python
#calcule de la fermeture puis correction 
REF_depart = [364387.872 , 376197.365 ] #364387,872 376197,365 364387,872 376197,365
DEP = [364367.943, 376228.353]   #364367,943 376228,353 364367,943 376228,353
REF_arriv = [364367.943, 376228.353]
ARRIV = [364300.889,376147.112]
try :
    G_arriv = gisement(ARRIV[0] , ARRIV[1],REF_arriv[0],REF_arriv[1])
    #G_depart = gisement(DEP[0] ,DEP[1] , REF_depart[0] , REF_depart[1]  )
    G_depart = gisement( REF_depart[0] , REF_depart[1],DEP[0] ,DEP[1] )
    n = len(angles_finales)
    gisements = np.zeros(n+1) 
    gisements[0] = G_depart 
    for  i in range(1,n+1 )   : 
        gisements[i] = roundd(gisements[i-1] + 200 + angles_finales[i-1] )
except ValueError : 
    print("coordonnee non valide ")
print(G_depart) 
print(gisements)  
print(G_arriv)
f = G_arriv - gisements[-1]
print(f)
#correction des angles 
angles_compensee = np.zeros(n)
for i in range(n) :
    angles_compensee[i] = angles_finales[i] + f / n
print(angles_finales)
print(angles_compensee)
```
